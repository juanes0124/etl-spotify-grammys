```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
print("Librerías cargadas correctamente")
```

```python
spotify = pd.read_csv('../data/spotify_dataset.csv', index_col=0)
grammy = pd.read_csv('../data/the_grammy_awards.csv')

print("Spotify shape:", spotify.shape)
print("Grammy shape:", grammy.shape)
```

```python
print("=== SPOTIFY - Tipos de datos ===")
print(spotify.dtypes)
print("\n=== Valores nulos ===")
print(spotify.isnull().sum())
print("\n=== Duplicados:", spotify.duplicated().sum(), "===")
print("\n=== Estadísticas básicas ===")
spotify.describe()
```

```python
print("=== GRAMMY - Tipos de datos ===")
print(grammy.dtypes)
print("\n=== Valores nulos ===")
print(grammy.isnull().sum())
print("\n=== Duplicados:", grammy.duplicated().sum(), "===")
print("\n=== Años disponibles ===")
print(sorted(grammy['year'].unique()))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Spotify - Distribución de Variables', fontsize=16)

spotify['popularity'].hist(ax=axes[0,0], bins=30, color='green')
axes[0,0].set_title('Popularidad')

spotify['danceability'].hist(ax=axes[0,1], bins=30, color='blue')
axes[0,1].set_title('Danceability')

spotify['energy'].hist(ax=axes[1,0], bins=30, color='orange')
axes[1,0].set_title('Energy')

spotify['track_genre'].value_counts().head(10).plot(kind='bar', ax=axes[1,1], color='purple')
axes[1,1].set_title('Top 10 Géneros')

plt.tight_layout()
plt.savefig('../notebooks/spotify_eda.png')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Grammy Awards - Análisis', fontsize=16)

grammy['category'].value_counts().head(10).plot(kind='barh', ax=axes[0], color='gold')
axes[0].set_title('Top 10 Categorías')

grammy['year'].value_counts().sort_index().plot(kind='line', ax=axes[1], color='red')
axes[1].set_title('Nominaciones por Año')

plt.tight_layout()
plt.savefig('../notebooks/grammy_eda.png')
plt.show()
```
